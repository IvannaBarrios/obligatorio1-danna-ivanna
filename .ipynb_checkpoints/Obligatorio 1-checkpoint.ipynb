## Ejercicio 1

Una máquina produce piezas, las cuales pueden ser defectuosas con probabilidad $p$. Considere un sistema de control de calidad que funciona en dos fases:

 * Fase A: Para cada pieza producida, se sortea al azar si se inspecciona con probabilidad $r$.
 * Fase B: Si se detecta una pieza defectuosa, se pasa a esta fase, y se inspeccionan todas las piezas producidas hasta obtener una racha de $N$ piezas normales, pasando luego a fase $A$.
 
Considere el estado $X(t)=$ "cant. de piezas consecutivas normales", con el estado $N$ representando que se encuentra en fase A, y $0,\ldots, N-1$ representando la racha de la fase $B$.

1. Modele el sistema anterior como una Cadena de Markov de tiempo discreto.
2. Discuta la irreducibilidad, periodicidad y ergodicidad de la cadena en función de $r,p,N$.
3. Calcule la distribución régimen en los casos que la cadena sea ergódica.
4. Calcule la proporción de piezas inspeccionadas efectivamente, como función de $r,p,N$.
5. Para $p=0.1$, $N=10$, estime $r$ numéricamente para que se inspeccione un $50\%$ de las piezas.
6. Simule la cadena de Markov para los valores anteriores y verifique que la distribución de régimen calculada es correcta.

### Parte 1

De la letra se obtiene:
- La probalidad de pieza defecuosa es p
- Cada pieza tiene probabiliad r de ser inspeccionada
- Si una pieza es defectuosa, se inspeccionan todas las piezas producidas hasta obtener N piezas normales

El siguiente diagrama representa los estados como cadena de Markov de tiempo discreto:

![Ejemplo](./Imagenes/Obli1_1.jpg)

En el diagrama se puede observar que:
- La fase A está representada como el estado N, donde se juntan dos premisas, se inspecciona la pieza con probabilidad r y además es defectuosa con probabilidad p, lo cual conlleva que para pasar a la fase B se tiene probabilidad rp.

Por lo tanto, para mantenerse en la fase A tiene probabilidad 1-rp.

La fase B comienza con el estado 0, chequeando que las piezas sean normales con probabilidad 1-p hasta el estado N donde se vuelve a la fase A.
Si durante el transcurso se detecta una pieza defectuosa, se vuelve a contar desde cero con probabilidad p.

### Parte 2

### Irreducibilidad
Para que la cadena sea irreducible todos los estados deben estar conectados.
Por lo tanto, para que esto ocurra, en el diagrama anterior se tiene que cumplir:

Si p = 0 no es irreducible ya que termina siempre en N. 

Si p = 1 no es irreducible ya que termina siemre en 0.

Si 0 < p < 1 y r $\neq$ 0 es irreducible.

### Periodicidad 

El criterio de periodicidad indica que un estado es periódico cuando puedo volver a él en múltiplos del mismo número mayor que 1.

Ademas si la probabilidad de volver a un estado en un paso es positiva, es aperiódico, y por consiguiente la cadena va a ser toda aperiódica por ser irreducible. 

Condiciones para el estado 0 y el estado N para que sea aperiódica:

  $p_{NN}\neq 0\quad$ *o* $\quad p_{00}\neq 0$


### Ergodicidad

La cadena será ergodica si las condiciones de irreducibilidad e aperiocidad son cumplida. Adicionalmnete, como la cadena es finita, se cumple el criterio de ergodicidad para cadenas finitas.

Sí es ergódica, tendrá un régimen bien definido y permite resolver las ecuaciones de régimen.

### Parte 3

La matriz de trancision de la cadena es:

$$P = \begin{pmatrix}
p & 1-p & 0 & 0 & \cdots & 0\\
p & 0 & 1-p & 0 & \cdots & 0\\
p & 0 & 0 & 1-p & \cdots & \vdots\\
\vdots & \vdots & \vdots & \vdots & \ddots & 0\\
\vdots & \vdots & \vdots & \vdots & \ddots & 1-p\\
rp & 0 & \cdots & 0 & 0  & 1-rp
\end{pmatrix}$$

Las ecuaciones seran:

$$\begin{align}
\pi_0 &= p(\pi_0 +\pi_1 +\pi_2+ \cdots + \pi_{N-1}+r\pi_N)\\
\pi_1 &= \pi_0(1-p) \\
\pi_2 &= \pi_1(1-p) \\
\pi_3 &= \pi_2(1-p) \\
&\vdots\\
\pi_{N-1} &= \pi_{N-2}(1-p) \\
\pi_{N} &= \pi_{N-1}(1-p)+\pi_{N}(1-rp)
\end{align}$$

Reduciendose a:
$$\begin{align}
\pi_j &= \pi_{j-1}(1-p) \quad & 1<j<N-1\\
\end{align}$$

               


## Ejercicio 2

Se considera un sistema de fila de espera al que llegan trabajos como un proceso de Poisson de intensidad $\lambda$. El sistema consta de 2 servidores. Cada trabajo es atendido por un único servidor y demora un tiempo $\exp(\mu)$ en servirse. Se plantean 3 políticas de operación:

* Política A: se mantiene una cola separada para cada servidor, cada cliente que llega es enviado al azar a una de ellas con igual probabilidad.

* Política B: se mantiene una cola separada para cada servidor, cada cliente que llega se envía a la *cola más corta* (sorteando en caso de empate).

* Política C: se mantiene una única cola para ambos servidores. Cada servidor que queda libre toma un trabajo de la cola, si hay.

1. Analice las tres políticas y discuta por qué la política B debería ser más eficiente que la A, y la C aún más que la B.

2. Muestre que en la política A, cada cola puede analizarse como una cola $M/M/1$. Calcule la cantidad media de clientes en el sistema y el retardo medio que ven los trabajos.

3. Muestre que la política C puede analizarse como una cola $M/M/2$ y dé una expresión para su distribución invariante.

4. Para la política $B$, considere como estado $X=(X_1,X_2)$ la cantidad de clientes en cada cola. Esboze el diagrama de estados de la política (no resolver la distribución, no tiene solución explícita!)

5. Simule el sistema en los tres casos anteriores para $\lambda=1.8$, $\mu=1.0$. Verifique las partes 2 y 3 para los sistemas A y C. El sistema B: ¿a cuál se parece más?
